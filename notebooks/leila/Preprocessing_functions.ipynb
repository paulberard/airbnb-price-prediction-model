```python
### Imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Loading data 
train_data = pd.read_csv('/Users/leilaberrada/Desktop/ml-project/data/train_airbnb_berlin_preprocessed.csv' , delimiter = ';')
train_data.head()
```

### Imputing neighbourhood

```python
## Fonctions
def get_closer_latitude_longitude(row):
    if pd.isnull(row['neighbourhood']):
        a=row['neighbourhood']
        print(f'inital value : {a}')
        latitude = row['Latitude']
        longitude = row['Longitude']
        train_data['distance'] = train_data.apply(lambda x: np.sqrt((x['Latitude'] - latitude)**2 + (x['Longitude'] - longitude)**2), axis=1)
        train_data['distance'] = train_data.apply( lambda x:1000 if x['distance'] == 0.0 or pd.isnull(x['neighbourhood']) else x['distance'], axis=1)
        row['neighbourhood'] = train_data[train_data['distance'] == train_data['distance'].min()]['neighbourhood'].values[0]
        b=row['neighbourhood']
        print(f'imputed value : {b}')
        
        return row['neighbourhood']
        
    else:
        return row['neighbourhood']
```

```python
train_data['neighbourhood'] = train_data.apply(lambda x: get_closer_latitude_longitude(x), axis=1)
train_data = train_data.drop(['distance'], axis=1)
train_data.head()
print(train_data['neighbourhood'].unique())
```

### Imputing Postal Code

```python
def get_closer_latitude_longitude_PC(row):
    if pd.isnull(row['Postal Code']):
        a=row['Postal Code']
        print(f'inital value : {a}')
        latitude = row['Latitude']
        longitude = row['Longitude']
        train_data['distance_PC'] = train_data.apply(lambda x: np.sqrt((x['Latitude'] - latitude)**2 + (x['Longitude'] - longitude)**2), axis=1)
        train_data['distance_PC'] = train_data.apply( lambda x:1000 if x['distance_PC'] == 0 or pd.isnull(x['Postal Code']) else x['distance_PC'], axis=1)
        row['Postal Code'] = train_data[train_data['distance_PC'] == train_data['distance_PC'].min()]['Postal Code'].values[0]
        b=row['Postal Code']
        print(f'imputed value : {b}')
        return row['Postal Code']
    else:
        return row['Postal Code']
```

```python
train_data['Postal Code'] = train_data.apply(lambda x: get_closer_latitude_longitude_PC(x), axis=1)
train_data = train_data.drop(['distance_PC'], axis=1)
train_data.head()
print(train_data['Postal Code'].unique())
```
